```python
import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])
```

```python
print(f"The Shape of the Matrix : {data.shape}")
print(f"\nThe Mean temperature per station : {np.mean(data)}")
```

```python
res = data > 28.0
print(f"The temperature above 28.0 is : {data[res]}")
```

```python
normalized = (data - data.min()) / (data.max() - data.min())
print(normalized:.2f)
```
