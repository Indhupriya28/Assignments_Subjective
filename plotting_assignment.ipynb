```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
print(tips.head())
```

```python
import matplotlib.pyplot as plt

sumvalue = tips.groupby("day")["tip"].sum()

plt.bar(sumvalue.index,sumvalue.values)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.figure(figsize=(8,5))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.histplot(data=tips, bins=20, kde=True)
plt.title("Distribution of Tip Amounts")
plt.show()

```

The graph shows most of the tips are small, data are skewed right side. KDE shows the smooth visualization of the pattern

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Total Bill vs Tip Amount")
plt.show()
```

yes, higher the total_bill ,higher the tips and vice-versa. also I could see on saturday higher the bill amount
