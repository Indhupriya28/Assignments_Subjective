Import pandas to convert the data into dataframe.
creating a student data and
converting to dataFrame.


```python
import pandas as pd
data = {
    "StudentID" : [101,102,103,104,105,106,107,108,109,110],
    "Name" : ['Alice','Bob','Charlie','Diana','Ethan','Fiona','George','Hannah','Ivan','Julia'],
    "Age" :[20,21,19,22,20,21,23,19,22,20],
    "Marks" : [88,75,92,65,78,85,70,95,60,80],
    "Grade" : ['A','B','A','C','B','A','B','A','C','B']
}

df_data = pd.DataFrame(data)
print(df_data)
```

to_csv --> converts the data from dataframe into csv format and saved in students.csv file. 
index = false -> remove the index value for the data

```python
df_data.to_csv("students.csv",index=False)
print ("Successfully created the csv file")
```

read_csv : Used to read the csv data and stored in new dataframe

```python
df= pd.read_csv("students.csv")
print(df)
```

Head(3) fetch first 3 data

```python
print(df.head(3))
```

describe -> produce the statistical data for all numerical columns . It includes count, mean, std, min,max

```python
print(df.describe())
```

To see the shape of the dataframe . It provide count of rows and columns

```python
print (f"The shape of the student dataframe is {df.shape}")
```

dtypes : define the datatype of the each column present in the dataframe

```python
print(f" DataTypes :")
print(df.dtypes)
```

Info() -> provide the detailed structural information of the dataset(all columns). Includes the datatype, missing values

```python
print(df.info())
```
