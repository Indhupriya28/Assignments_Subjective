```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}
```

```python
import pandas as pd

student_df = pd.DataFrame(students_data)
enroll_df = pd.DataFrame(enrollments_data)
score_df = pd.DataFrame(scores_data)

```

```python
# Display null value count and percentage for each column
# Fill missing 'city' values with 'Unknown'
# Drop rows where 'name' is missing
print("=== TASK 1: Data Preparation ===")

print(f"Original Students DataFrame:")
print(student_df)

nullCount = student_df.isnull().sum()
nullPercent = nullCount / len(student_df) *100

print(f"Null Value Analysis: \n")
for col in student_df.columns:
  print(f"Column: {col}, Nulls: {nullCount[col]} ({nullPercent[col]:.2f})")

student_df.fillna({'city':'Unknown'}, inplace=True)
len_before_drop = len(student_df)
student_df.dropna(subset=['name'],inplace=True)
len_after_drop = len(student_df)
print("Cleaned Students DataFrame: ",{len_after_drop})
print(student_df)
```

TASK-2 MULTIPLE JOIN OPERATION

```python
#Inner Join

result = pd.merge(student_df,enroll_df, on='student_id',how='inner')
print(f"\nInner Join Result: ({len(result)} rows)")
print(result)
```

Total of 4 records are fetch as a result.
Inner join fetch the record from the both the table which has common id in it.
Student_id (104, 106, 107) is excluded because these id doesn't have matching data in the enrollment table

```python
#Left Join

result = pd.merge(student_df,enroll_df, on='student_id',how='left')
print(f"\nLeft join result: ({len(result)} rows)")
print(result)
```

total 6 rows are present
students 104,106,107 has NaN, because these doesn't have matching student_id in the course table (with respective to Student table)

```python
#Right Join

result = pd.merge(student_df,enroll_df, on='student_id',how='right')
print(f"\nRight join result: ({len(result)} rows)")

print(result)
```

Student_id : 103, 108, 109 doesnt gave name in the student table. These data are not present in the enrollment table

```python
result = pd.merge(student_df,enroll_df, on='student_id',how='outer',indicator=True)
print(f"\nOuter join result: ({len(result)} rows)")

print(result)
missing_row = result['name'].isnull() | result['course_name'].isnull()
print(result[missing_row])
```

Task 3: Lookup Operation and Automation

```python
scores = dict(zip(scores_data['student_id'],scores_data['exam_score']))
student_data = student_df.copy()
student_data['exam_score'] = student_data['student_id'].map(scores)

print(student_data)
```

```python
student_record = pd.merge(student_df,score_df,on='student_id',how='left')

print(student_record)
```

```python
%timeit  pd.merge(student_df,score_df,on='student_id',how='left')

%timeit  student_data['student_id'].map(scores)
```

Since we are add one column .map() is more efficient that merge.
Merge can be used when more than 2 columns need to be added. By using merge, this will create a new dataframe. causing more time and memory when compared to map()

SIMPLE AUTOMATION

```python
def auto_merge(df1, df2, join_type, key_column):
  print(pd.merge(df1,df2,on=key_column,how=join_type))
  print("\n")


auto_merge(student_df,score_df,'outer','student_id')
auto_merge(student_df,score_df,'right','student_id')
```
