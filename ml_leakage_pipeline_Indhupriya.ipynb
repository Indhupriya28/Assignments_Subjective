```python
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#dataset
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)

#scaling before data split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#data split by deafult test_size is 0.25
x_train,x_test,y_train, y_test = train_test_split(X_scaled,y,test_size=0.3, random_state=42)

#Build a model
model = LogisticRegression()
#train a model
model.fit(x_train,y_train)

#predict the model
model.score(x_test,y_test)
```

Here, scaling happens for test data also. So mean and standard deviation is calculation includes test data. When actual test data is provided for scoring , it wont be the right one. even though the accuracy is ~85%. This model fails with real data prediction

```python
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline

X,y = make_classification(n_samples =1000, n_features=10, random_state=42)

x_train, x_test, y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)

#pipeline created with scaler and model
pipe = Pipeline([
    ('scaler',StandardScaler()),
     ('model',LogisticRegression())
])

#cross fold evaluation with 5 fold
scores = cross_val_score(pipe,x_train,y_train,cv=5)

print(f"Score for each fold: {scores.round(2)}")
print(f"Mean Score : {scores.mean().round(2)}")
print(f"Standard deviation : {scores.std().round(2)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
X,y = make_classification(n_samples =1000, n_features=10, random_state=42)

x_train, x_test, y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)

result =[]
#model = DecisionTreeClassifier(max_depth=4,random_state=42)
for depth in [1,5,20]:
  model = DecisionTreeClassifier(max_depth=depth, random_state=42)
  model.fit(x_train,y_train)
  train_accuracy=model.score(x_train, y_train)
  test_accuracy= model.score(x_test, y_test)

  result.append([depth, train_accuracy,test_accuracy])

df = pd.DataFrame(result, columns=['Depth','Train_Accuracy','Test_Accuracy'])
print(f"{df}")
```

Among 3 depth, "max_depth = 5" is a good fit.  The max_depth with "1" though it gives nearly same accuracy value for the train and test data. there could be possibility that all data falls in one side of the tree (simple model). Underfitting.
Depth with "20" - overfitting.Huge difference between train and test data.
