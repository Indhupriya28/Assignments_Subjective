```python
import pandas as pd
import numpy as np

data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                   111, 112, 113, 114, 115, 101, 107, 118, 119, 120],
    'age': ['25', '34', None, '45', '29', None, '38', '52', '27', '41',
            '33', 'unknown', '48', '26', '35', '25', '38', '31', None, '44'],
    'weight': ['70', '65', '80', None, '75', None, '68', '90', '72', '85',
               '78', None, '82', '69', 'N/A', '70', '68', '74', None, '88'],
    'blood_pressure': [120, 130, None, 140, 125, None, 135, None, 118, 145,
                      128, None, 138, 122, None, 120, 135, 126, None, 142],
    'medication': ['Aspirin', 'Metformin', 'Lisinopril', None, 'Aspirin',
                   'Metformin', 'Lisinopril', 'Aspirin', None, 'Metformin',
                   'Lisinopril', 'Aspirin', None, 'Metformin', 'Aspirin',
                   'Aspirin', 'Lisinopril', 'Metformin', 'Aspirin', None],
    'insurance_provider': ['Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None,
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', 'Blue Cross',
                          'Aetna', None, 'UnitedHealth', 'Blue Cross', 'Aetna',
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None]
}

df = pd.DataFrame(data)
```

TASK 2 : DATATYPE CONVERSION

```python
import pandas as pd

# Count missing values before conversion
missing_before_age = df['age'].isna().sum()
missing_before_weight = df['weight'].isna().sum()

#Convert age column to numeric
df['age'] = pd.to_numeric(df['age'], errors='coerce')

#Convert weight column to numeric
df['weight']= pd.to_numeric(df['weight'], errors='coerce')

# Count missing values before conversion
missing_after_age = df['age'].isna().sum()
missing_after_weight = df['weight'].isna().sum()

# how many NEW missing values were created after conversion
new_missing_age = missing_after_age - missing_before_age
new_missing_weight = missing_after_weight - missing_before_weight

print("New missing values in age:", new_missing_age)
print("New missing values in weight:", new_missing_weight)

# Convert insurance_provider to category type
df['insurance_provider'] = df['insurance_provider'].astype('category')

# Verify conversions
print(df.dtypes)
```

TASK 3 HANDLE MISSING VALUES

```python
df.info()
```

```python
#age: Fill with median  
df['age'] = df['age'].fillna(df['age'].median())

#weight: Fill with median  
df['weight'] = df['weight'].fillna(df['weight'].median())

#blood_pressure: Fill with median 
df['blood_pressure'] = df['blood_pressure'].fillna(df['blood_pressure'].median())

#medication: Fill with mode 
df['medication'] = df['medication'].fillna(df['medication'].mode()[0])

#insurance_provider: Fill with constant value 'Unknown'

#df['insurance_provider'] = df['insurance_provider'].fillna('Unknown') --> gives error because, type already in Categorial . Pandas doesn't allow to 
# fill the value (Unknown here) which is not present in category.
# so convert to Str and then fill NA with Uknown
df['insurance_provider'] = df['insurance_provider'].astype(str).fillna('Unknown').astype('category')

print(df.isnull().sum())
```

TASK 4: HANDLE DUPLICATES

```python
#to view the duplicates

print(df[df.duplicated(keep=False)])
```

```python
print(df.duplicated(subset=['patient_id']))
```

```python
print(f"Shape before dropping duplicates {df.shape}")
df = df.drop_duplicates(subset=['patient_id'], keep='first')
print(f"\nShape after dropping duplicates {df.shape}")
```

TASK 5 Complete workflow with Verification

```python
#creating new DF 
df_clean = df.copy()

#Before 
shape_before = df_clean.shape
dtypes_before = df_clean.dtypes
missing_before = df_clean.isna().sum()

age_missing_before = df_clean['age'].isna().sum()
weight_missing_before = df_clean['weight'].isna().sum()
bp_missing_before = df_clean['blood_pressure'].isna().sum()
medication_missing_before = df_clean['medication'].isna().sum()
insurance_missing_before = df_clean['insurance_provider'].isna().sum()


duplicates_before = df.duplicated().sum()

#Type conversion

df_clean['age'] = pd.to_numeric(df_clean['age'],  errors='coerce')
df_clean['weight'] = pd.to_numeric(df_clean['weight'], errors = 'coerce')


df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())
df_clean['weight'] = df_clean['weight'].fillna(df_clean['weight'].median())
df_clean['insurance_provider'] = df_clean['insurance_provider'].fillna('Unknown')

df_clean['insurance_provider'] = df_clean['insurance_provider'].astype('category')

# Remove duplicates
df_clean = df_clean.drop_duplicates()

# --- Step 4: Record AFTER stats ---
shape_after = df_clean.shape
missing_after = df_clean.isna().sum()
duplicates_after = df_clean.duplicated().sum()
dtypes_after = df_clean.dtypes

report = {
    "Shape before": shape_before,
    "Shape after": shape_after,
    "Missing values before": missing_before,
    "Missing values after": missing_after,
    "Duplicates before": duplicates_before,
    "Duplicates after": duplicates_after,
    "Data types before": dtypes_before,
    "Data types after": dtypes_after
}

# Display neatly
for key, value in report.items():
    print(f"\n{key}:\n{value}")
```
